```python
cal['event_type_1'].replace(to_replace=np.nan, value="nil", inplace=True)
```

```python
cal['event_type_2'].replace(to_replace=np.nan, value="nil", inplace=True)
```

```python
# create a column with year and month info

yrmth =[]
for x in range(len(cal)) :
    temp = str(cal['year'][x])+'-'+'{:02.0f}'.format(cal['month'][x])
    yrmth.append(temp)

cal['yrmth'] = yrmth
cal
```

```python
cal['event1'] = cal['event_name_1'].apply(lambda x : 1 if x != 'nil' else 0) 
```

```python
cal['event2'] = cal['event_name_2'].apply(lambda x : 1 if x != 'nil' else 0) 
```

```python
## create a column to capture the dates with event_name_1 and event_name_2 not nil ie having 2 events at the same date
events =[]
for i in range(len(cal['event1'])) :
    event_ct = cal['event1'][i]+cal['event2'][i]
    if event_ct > 1 :
        count = 1
    else            : 
        count = 0
    events.append(count)

cal['events'] = events
cal[cal['events']==1]
```

```python
## replace event_name_1 = nan with Thanksgiving_Oth for 1 and 2 days after Thanksgiving day 

for x in range(len(cal)) :
    if cal.event_name_1[x] == 'Thanksgiving' :
        y = x + 1
        z = x + 2
        cal.event_name_1[y] = 'Thanksgiving_Oth'
        cal.event_name_1[z] = 'Thanksgiving_Oth'
    else :
        pass

    
```

```python
cal['is_eventR']=cal['event_type_1'].apply(lambda x : 1 if x == 'Religious' else 0)
cal['is_eventN']=cal['event_type_1'].apply(lambda x : 1 if x == 'National' else 0)
cal['is_eventC']=cal['event_type_1'].apply(lambda x : 1 if x == 'Cultural' else 0)
cal['is_eventS']=cal['event_type_1'].apply(lambda x : 1 if x == 'Sporting' else 0)
```

```python
cal['Xmas'] = cal['event_name_1'].apply(lambda x : 1 if x == 'Christmas' else 0)
cal['TG'] = cal['event_name_1'].apply(lambda x : 1 if ((x == 'Thanksgiving') | (x == 'Thanksgiving_Oth')) else 0)
cal['NY'] = cal['event_name_1'].apply(lambda x : 1 if x == 'NewYear' else 0)
```

```python
cal['LD'] = cal['event_name_1'].apply(lambda x : 1 if x == 'LaborDay' else 0)
cal['SB'] = cal['event_name_1'].apply(lambda x : 1 if ((x == 'SuperBowl')) else 0)
```

```python
## create eve of holidays column 
cal['eve'] = [0 for x in range(cal.shape[0])] 

for x in range(len(cal)) :
    if x == 0 :
        cal.eve[x] = 0
    elif (cal.event_name_1[x] != 'nil') & (cal.event_name_1[x-1] != 'nil'):
        cal.eve[x-1] = 0
        cal.eve[x] = 0  
    
    elif (cal.event_name_1[x] != 'nil') & (cal.event_name_1[x-1] == 'nil'):
            cal.eve[x-1] = 1
            cal.eve[x] = 0
            
    else :
        cal.eve[x] = 0
```
